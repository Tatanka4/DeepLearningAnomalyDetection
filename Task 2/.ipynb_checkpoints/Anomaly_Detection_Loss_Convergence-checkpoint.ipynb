## **Data Preparation**

##### **Patient series split**
Nella cella sottostante verrÃ  ricreato l'array con le sequenze per ogni paziente dove perÃ² le series dei pazienti con dei valori di "Heartrate" a -1 verranno considerate come due series: una a sinistra dei valori di "-1" ed una a destra dei valori di "-1"

```python
new_data = []

filter_hr = train_data_loaded['heartRate'] == -1
for id in id_patients:
  
  filter_id = train_data_loaded['patient'] == id
  
  series = train_data_loaded.where(filter_id).dropna()
  
  if id in ids_missing_hr:
    missing_hr_row = train_data_loaded.where(filter_id & filter_hr).dropna()

    left_series = train_data_loaded[series.index[0]:missing_hr_row.index[0]]
    right_series = train_data_loaded[missing_hr_row.index[-1]+1:series.index[-1]]

    if (len(left_series)) != 0:
        new_data.append(left_series)
    
    if (len(right_series)) != 0:
        new_data.append(right_series)
    #pass
  else:
    new_data.append(series)
```

##### **Data Windowing**
Le due funzioni sottostanti verranno utilizzate rispettivamente dal training set e dal test set. La prima funzione tornerÃ  un'array di sequenze raccolte da ogni paziente. La seconda funzione tornerÃ  un'array di sequenze di quel solo paziente

```python
def data_windowing(data, lookback, step):
    
    patient_series = []

    for patient_id_index in range(0,len(data)):
        for i in range(0, len(data[patient_id_index]) - lookback - 1, step):
            patient_series.append(data[patient_id_index][i: i + lookback])
      
    patient_series = np.array(patient_series)
    patient_series = patient_series.reshape((patient_series.shape[0], patient_series.shape[1], 4))
    return patient_series

def data_windowing_test(data, lookback, step):
    
    patient_series = []
    
    for i in range(0, len(data) - lookback - 1, step):
        patient_series.append(data[i: i + lookback])
      
    patient_series = np.array(patient_series)
    patient_series = patient_series.reshape((patient_series.shape[0], patient_series.shape[1], 4))
    return patient_series
```

```python
win_size = 30
win_shift = 1

x_train = data_windowing(train_data, win_size, win_shift)
x_val = data_windowing(val_data, win_size, win_shift)

seq_test = []
for p in test_data:
    seq_test.append(data_windowing_test(p, win_size, win_shift))
```

#### *Metrics*

L'indice di "mae" ottenuto dalla predizione su ogni riga di ogni sequenza di ogni paziente Ã¨ "compresso" come somma in modo da non avere risultati doppioni da queste predizioni. 

```python
patient_mae_per_row = {}
for i in range(0, len(p_errors)):
    patient_mae_per_row[i] = np.zeros(len(p_errors[i]))
    for j in range(0, len(p_errors[i])- win_size - 1):
        for k in range(0, len(p_errors[i][j])):
            patient_mae_per_row[i][j+k] += p_errors[i][j][k]
```

```python
for i in range(0, len(p_errors)):
    for j in range(0, len(p_errors[i])):
        if j < win_size - 1:
            patient_mae_per_row[i][j] = patient_mae_per_row[i][j]/(j+1)
        elif j > len(p_errors[i]) - win_size:
            patient_mae_per_row[i][j] = patient_mae_per_row[i][j]/(len(p_errors[i])-j)
        else:
            patient_mae_per_row[i][j] = patient_mae_per_row[i][j]/win_size
```

## *Evaluation* 

Per ogni paziente viene mostrato il grafico con l'andamento dell'errore calcolato dalle predizioni. Su di ognuno di questi andamenti vengono considerati come "off period" quei valori di errori che si avvicinano al 95% dell'errore massimo.   

```python
tresholds = []
for p in range(0, len(patient_mae_per_row)):
    er_max = max(patient_mae_per_row[p])
    print(er_max)
    print("----")
    tr = ((er_max) / 10 * 9.5)
    tresholds.append(tr)
    plt.plot(range(len(patient_mae_per_row[p])), patient_mae_per_row[p],  label='Error')
    plt.axhline(tr, color='r')
    plt.title('Error') 
    plt.xlabel('Record')
    plt.ylabel('Loss') 
    plt.legend()
    plt.show()
```

##### Saving result
I momenti di off-period vengono salvati in un file .csv contenente le colonne di "patient" (id del paziente), "tsDate" (momento e ora dell'offperiod) e "off" (stringa "True" che attesta questo avvenimento)

```python
test_data_copy

f = open("off_period.csv", "w")
f.write("patient, tsDate, off\n")

for p in range(len(patient_mae_per_row)):
    for er in range(len(patient_mae_per_row[p])):
        if patient_mae_per_row[p][er] > tresholds[p]:
            row = str(test_data_copy[p].iloc[er]['patient']) + "," + str(test_data_copy[p].iloc[er]['tsDate']) + ", True\n"
            f.write(row)
f.close()
```
